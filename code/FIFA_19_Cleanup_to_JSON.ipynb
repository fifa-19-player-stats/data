### "camelCase" the columns

```python
def camelCase(st):
    output = ''.join(x for x in st.title() if x.isalnum())
    return output[0].lower() + output[1:]
  
def ccCols(df):
  new_cols = []
  for col in df.columns:
    new_col = camelCase(col)
    new_cols.append(new_col)
  return new_cols
    
```

```python
df_raw.columns= ccCols(df_raw)

df_raw.head()
```

### Filling in the Club nulls

I will fill in the Club nulls using information from the Club Logo column.

```python
df['clubLogo'].nunique(), df['club'].nunique()
```

```python
# df['club'].fillna('No Club Listed', inplace = True)

df['club'].isnull().sum()
```

```python
def strip_url(url):
  piece = url.split("/")[-1:]
  piece = piece[0].replace('.png','')

  return piece

strip_url('https://cdn.sofifa.org/teams/2/light/110955.png')
```

```python
df['clubNum'] = df['clubLogo'].apply(strip_url)

df.head()
```

**Creating a dictionary to map values into Club**

```python
from collections import defaultdict

d = defaultdict(list)

for idx, row in df.iterrows():
  d[row['clubNum']].append(row['club'])
    
# unique_items = set(x for y in d.values() for x in y)

# unique_items

# d = [{k: v} for k, v in d.items()}

d
```

```python
d2 = [] # List of tuples

for key, value in d.items():
  new_value = 0
  if max(set(value), key=value.count) == np.nan:
    new_value = min(set(value), key=value.count)
  else:
    new_value = max(set(value), key=value.count)
  d2.append((key,new_value))

# d['10'][0]

print(len(d2))

d2.sort(key=lambda tup: tup[0])

d2
```

```python
d3 = []

for i in range(len(d2)):
  if d2[i][1] == "No Club Listed":
    d3.append(d2[i][0])
    
d3
```

```python
d2 = dict(d2)

d2
```

**Filling in the correct values**

```python
for key, value in d2.items():
  if key == '103':
    d2[key] = 'Cameroon'
  elif key == '108':
    d2[key] = 'Ivory Coast'
  elif key == '111':
    d2[key] = 'Egypt'
  elif key == '140':
    d2[key] = 'South Africa'
  elif key == '159':
    d2[key] = 'DSC Arminia Bielefeld'
  elif key == '195':
    d2[key] = 'Australia'
  elif key == '198':
    d2[key] = 'New Zealand'
  elif key == '24':
    d2[key] = 'Iceland'
  elif key == '37':
    d2[key] = 'Cameroon'
  elif key == '40':
    d2[key] = 'Russia'
  elif key == '53':
    d2[key] = 'Bolivia'
  elif key == '58':
    d2[key] = 'Paraguay'
  elif key == '60':
    d2[key] = 'Uruguay'
  elif key == '61':
    d2[key] = 'Venezuela'
 
```

```python
d2
```

```python
df['club'] = df['clubNum'].map(d2)

df['club'] = df['club'].astype(str)

df.isnull().sum()
```

**Club is now fixed**

**Converting the money columns into separate columns**

```python
money_cols = ['value', 'wage', 'releaseClause']

for col in money_cols:
  df[col].fillna('€0.0M', inplace=True)
  df[col].replace('€0', '€0.0M', inplace=True)
  
df.isnull().sum()
```

**Stripping the money columns into a currency symbol, a float and a multiplier**

```python
df['currency'] = df['value'].map(lambda x: x[0]).astype(str) # The money columns are all in Euros
df['valueAmt'] = df['value'].map(lambda x: x[1:-1]).astype(float)
df['valueUnit'] = df['value'].map(lambda x: x[-1]).astype(str)


df['wageAmt'] = df['wage'].map(lambda x: x[1:-1]).astype(float)
df['wageUnit'] = df['wage'].map(lambda x: x[-1]).astype(str)


# df['releaseClause'] = df['releaseClause'].astype(str) Fixed the nulls so not necessary.

df['releaseClauseAmt'] = df['releaseClause'].map(lambda x: x[1:-1]).astype(float)
df['releaseClauseUnit'] = df['releaseClause'].map(lambda x: x[-1]).astype(str)


df.head()
```

```python
df['valueEuro'] = df['valueAmt']
df['wageEuro'] = df['wageAmt']
df['releaseClauseEuro'] = df['releaseClauseAmt']

df['valueEuro'].loc[df['valueUnit'] == 'K'] = df['valueEuro']*1000
df['valueEuro'].loc[df['valueUnit'] == 'M'] = df['valueEuro']*1000000

df['wageEuro'].loc[df['wageUnit'] == 'K'] = df['wageEuro']*1000
df['wageEuro'].loc[df['wageUnit'] == 'M'] = df['wageEuro']*1000000

df['releaseClauseEuro'].loc[df['releaseClauseUnit'] == 'K'] = df['releaseClauseEuro']*1000
df['releaseClauseEuro'].loc[df['releaseClauseUnit'] == 'M'] = df['releaseClauseEuro']*1000000



df.head()
```
